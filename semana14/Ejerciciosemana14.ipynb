```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
h=0.0033
t=np.arange(1,1000,h)
#print(t)
v=np.zeros(len(t))
v[0]=8
x=np.zeros(len(t))
x[0]=500
w=5
#print(x)
#print(v)
```

```python
for i in range (1000):
    x[i+1]=x[i]+(h*v[i])
    v[i+1]=v[i]+(h*(-w**2*x[i]))
    #print(v[i])
    #print(w)
    plt.plot(t[i],v[i],'.')
    plt.plot(t[i],x[i],'.')
    #plt.plot(v[i],x[i],'o')
```

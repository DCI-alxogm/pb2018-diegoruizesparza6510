```python
#Se declara todo de una forma mas simple para ahorrar espacio
import numpy as np
import matplotlib.pyplot as plt
#Se da la instrcucci√≥n para que se muestre en el notebook
%matplotlib inline
```

```python
x=np.linspace(-10,10,100) #Se crea un arreglo de 100 elementos con un limite inferior de -10 y superior de 10 con un espacio igual entre cada elemento
x2=np.linspace(0,5,100) #Se crea un arreglo de 100 elementos con un limite inferior de 0 y superior de 5 con un espacio igual entre cada elemento
y=x**2*np.cos(x)-3*x**3 # Se trabaja sobre cada valor de x y el resultado se guarda en un arreglo del mismo tama√±no que el de x 
y2=x2**2*np.cos(x2)-3*x2**3
plt.plot(x,y) #se grafica x contra y
plt.plot(x2,y2)
```
